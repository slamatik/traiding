```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import numpy as np
import matplotlib as mpl
from datetime import datetime
```

```python
with open('data.pkl', 'rb') as f:
    data = pickle.load(f)
    
with open('yf_data.pkl', 'rb') as f:
    yf_data = pickle.load(f)
    
yf_tickers = ['msft', 'tsla', 'aapl', 'nio', 'pltr', 'adsk', 'clf']
```

```python
# example yf dataframe
yf_dict = {i: [] for i in yf_tickers}
for ticker in yf_tickers:
    yf_dict[ticker].append(yf_data[ticker]['operatingMargins'])
    yf_dict[ticker].append(yf_data[ticker]['revenueGrowth'])
    yf_dict[ticker].append(yf_data[ticker]['currentRatio'])
yf_df = pd.DataFrame(yf_dict, index=['operatingMargins','revenueGrowth','currentRatio']).T
```

```python
yf_df
```

```python
yf_df.sort_values(by='operatingMargins', axis=0, ascending=False)
```

```python
n_rows = len(yf_df)
n_cols = len(yf_df.columns)
size = 1 / n_cols - 0.1
x = np.arange(n_rows)

fig, ax = plt.subplots(figsize=(12, 8))
for col in range(n_cols):
    ax.bar(x + size * col, yf_df[yf_df.columns[col]], width=size, label=yf_df.columns[col])
plt.xticks(x, yf_df.index)
ax.legend()
```

```python
def plot_bar_chart(df, plot_type='', orientation='h', sort='', y_axis='l', date=datetime.today(), source='Bloomberg', plot=False):
    n_rows = len(df)
    n_cols = len(df.columns)
    size = 1 / n_cols - 0.1
    x = np.arange(n_rows)
    
    fig, ax = plt.subplots(figsize=(12, 8))
    for col in range(n_cols):
        ax.barh(x + size * col, df[df.columns[col]], height=size, label=yf_df.columns[col])
    plt.yticks(x, df.index)
    ax.legend()

plot_bar_chart(yf_df)
```
